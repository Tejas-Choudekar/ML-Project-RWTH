```python
# Since there aree lots of outlier, we used median to fill the missing values
df  = data.copy()
df = df.drop(columns="Attr37")
df.fillna(df.median(), inplace=True)


```

```python
miss_per = ((df.isnull().sum())*100/len(df["Attr1"]))
pd.DataFrame(miss_per).T

# No missing values
```

```python
# Outlier treatment using replacing outlier with NaN and then filling them with median

df_Y = df["class"]
variables = [c for c in df.columns if c!="class"]

for variable in variables:
    Q1 = df[variable].quantile([0.25]).values[0]
    Q3 = df[variable].quantile([0.75]).values[0]
    IQR = [Q1 - 1.5*(Q3-Q1) ,Q3 + 1.5*(Q3-Q1) ]
    median = float(df[variable].median())
    df.loc[df[variable] > IQR[1],variable] = np.nan
    df.loc[df[variable] < IQR[0],variable] = np.nan
    df.fillna(median, inplace=True)
    

df.describe()
    


```

```python
## Feature Selection: For this we are using RandomForestClassifier

from sklearn.model_selection import train_test_split

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestClassifier
```

```python
x = Bdata.drop('class', axis=1)
y = Bdata['class']
trainX, testX, trainY, testY = train_test_split(x, y, test_size = 0.25)
```

```python
clf = RandomForestClassifier(n_jobs=-1)

# Build step forward feature selection
feature_selector = SequentialFeatureSelector(clf,
           k_features=30,
           forward=True,
           floating=False,
           verbose=2,
           scoring='accuracy',
           cv=5)

features = feature_selector.fit(trainX, trainY)
filtered_features= trainX.columns[list(features.k_feature_idx_)]
filtered_features
```
