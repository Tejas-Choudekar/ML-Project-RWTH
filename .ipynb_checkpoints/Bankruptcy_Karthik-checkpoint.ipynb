### Creating 2 new dataFrames for Bankrupt and Not Bankrupt for imputation


```python
d_bankrupt = d.loc[d['class'] == "b'1'"]
d_not_bank= d.loc[d['class']=="b'0'"]
    
```

```python
#d_bankrupt.isna().value_counts()
d_bankrupt.isna().sum() #Gives missing values in each column for bankrupt firms
```

```python
d_not_bank.isna().sum()
```

```python
#Replacing missing data of bankrupt and non_bankrupt firms separately by medians of corresponding features
d_bankrupt.fillna(d_bankrupt.median(),inplace=True)
```

```python
d_bankrupt.isna().sum()
```

```python
d_not_bank.fillna(d_not_bank.median(),inplace=True)
```

```python
d_not_bank.isna().sum()
```
